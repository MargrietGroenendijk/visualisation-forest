<span style="color: #01807D; font-size: 2em;">**matplotlib**</span>

Data from https://www.esrl.noaa.gov/gmd/ccgg/trends/data.html

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
co2_mlo = pd.read_csv("data/co2_weekly_mlo.csv")

co2_mlo = co2_mlo.replace(-999.99, np.nan)

co2_mlo.head(5)
```

```python
fig, ax = plt.subplots()
ax.plot(co2_mlo['decimal'], co2_mlo['ppm'])
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(co2_mlo['decimal'], co2_mlo['ppm'],'-',color='#01807D', linewidth=2)

ax.set_title('Observed CO$_2$ at Mauna Loa',loc='right',fontsize=20)
ax.set_xlabel('year',fontsize=16)
ax.set_ylabel('CO$_2$ (ppmv)',fontsize=16)
ax.set_title('CO$_2$ at Mauna Loa',loc='right',fontsize=20)

plt.show()
```
