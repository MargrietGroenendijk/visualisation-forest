<span style="color: #A51B73; font-size: 2em; font-weight:bold">CO2 from ice cores</span>

Data from Barnola et al. Nature, 329, 408-414 (1987)

````python
import pandas as pd
%matplotlib inline

co2_ice = pd.read_csv("data/co2.csv")

s = co2_ice.plot(x='ice Age', y='CO2 ppmv')
````

<span style="color: #A51B73; font-size: 2em; font-weight:bold">CO2 at Mauna Loa</span>

Data from https://www.esrl.noaa.gov/gmd/ccgg/trends/data.html

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

co2_mlo = pd.read_csv("data/co2_weekly_mlo.csv")
co2_mlo = co2_mlo.replace(-999.99, np.nan)

fig, ax = plt.subplots()
ax.plot(co2_mlo['decimal'], co2_mlo['ppm'])
None
````

````python
fig, ax = plt.subplots()

ax.plot(co2_mlo['decimal'], co2_mlo['ppm'],'-',color='#01807D', linewidth=2)
ax.set_title('Observed CO$_2$ at Mauna Loa',loc='right',fontsize=20)
ax.set_xlabel('year',fontsize=16)
ax.set_ylabel('CO$_2$ (ppmv)',fontsize=16)
ax.set_title('CO$_2$ at Mauna Loa',loc='right',fontsize=20)
None
````

<span style="color: #A51B73; font-size: 2em; font-weight:bold">Comparison of data with seaborn</span>

````python
import seaborn as sns

sns_plot = sns.distplot(co2_mlo['ppm'].dropna())
sns_plot = sns.distplot(co2_ice['CO2 ppmv'].dropna())
fig = sns_plot.get_figure()
#fig.savefig('hist.png', dpi = 600)
````

````python
import pdvega

co2_mlo.vgplot.line(x='decimal', y='ppm', alpha=0.5)
````

````python
co2_mlo.vgplot.heatmap(x='yr', y='mon', C='ppm', gridsize=20)
````
